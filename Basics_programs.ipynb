```python
#recursion basics
def message(n):
 if n<0:
   return
 else:
   n=n-1
   print('welcome',n)
   message(n)
message(5)
```

```python
#display 10 to 1 using recursion

def Number(n):
  if n<1:
    return 1
  else:
    print(n)
    Number(n-1)
Number(10)
```

```python
# sum of n numbers using recursion

def Total(n):
   if n<1:
     return 1

   return n + Total(n - 1)

print(Total(5))
```

```python
#factorial using recursion
def factorial(n):
  if n<=1 or n==0:
    return 1
  else:
    return n * factorial(n-1)
fct = factorial(5)
print(fct)
```

```python
#fibonacci series using

def Fibonacci(n: int) -> int:
    if n == 0:
        return 0
    if n == 1:
        return 1
    return Fibonacci(n - 1) + Fibonacci(n - 2)

fibs = Fibonacci(6)
print(fibs)
```

```python
from io import FileIO
class File:
  def create(self):
    try:
      with open('hello.txt','w') as fl:

            fl.write('welcome to python session')
    except FileIO:
      print('File Error')
  def Reads(self):
    try:
      fle=open('hello.txt','r')
      print(fle.read())
    except File:
      print('Read error')
obj = File()
obj.create()
obj.Reads()
```
