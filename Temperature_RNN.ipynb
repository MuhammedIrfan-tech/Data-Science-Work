```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from tensorflow.keras.layers import SimpleRNN,Dense
from tensorflow.keras.models import Sequential
```

```python
data = pd.read_csv('temp.csv')
```

```python
datas=data.drop(columns='day').values
```

```python
#scale the data
scale = MinMaxScaler()
temp=scale.fit_transform(datas)
```

```python
#create 7 days sequence
x,y=[],[]
for i in range(len(temp)-3):
  x.append(temp[i:i+3])
  y.append(temp[i+3])
```

```python
x=np.array(x)
y=np.array(y)
```

```python
#create model
model = Sequential([
    SimpleRNN(units=30,activation='relu',input_shape=(3,1)),
    Dense(1)
])
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x,y,epochs=10)
```

```python
#predict new data
arr=np.array([15.2,15.5,15.7]).reshape(3,1)
temp = scale.transform(arr)
pred=model.predict(arr)
pred = scale.inverse_transform(pred)
print(pred[0][0])
```
