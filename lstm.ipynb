```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM
```

```python
#set the data
temps=np.array([20, 21, 19, 22, 23, 24, 25, 26, 27],dtype=float).reshape(-1,1)
```

```python
#create sequence with 3 samples
x=[]
y=[]
for i in range(len(temps)-3):
  x.append(temps[i:i+3])
  y.append(temps[i+3])
```

```python
x_train=np.array(x)
y_train=np.array(y)
```

```python
#create LSTM model

model = Sequential()
model.add(LSTM(40,activation='relu',input_shape=(3,1)))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python

model.fit(x_train,y_train,epochs=100,verbose=1)
```

```python
inp = np.array([20, 21, 19]).reshape(1, 3, 1)
prediction = model.predict(inp, verbose=0)

print("Predicted value:", np.floor(prediction[0][0]))
```
